```python
def best_premutation(arr1, i=0):
    if i == len(arr1):
        print("".join(map(str,arr1)))
        return
    
    for j in range(i, len(arr1)):
        arr1[i], arr1[j] = arr1[j], arr1[i]  # swap
        best_premutation(arr1, i+1)          # recurse
        arr1[i], arr1[j] = arr1[j], arr1[i]  # swap back

# ✅ Pass list instead of string
best_premutation([1,4,6,8,19])
```

```python

from langchain_community.document_loaders import TextLoader
from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import OllamaEmbeddings
from langchain_text_splitters import CharacterTextSplitter

loader = TextLoader("speech.txt")
document = loader.load()
text_splitter = CharacterTextSplitter(chunk_size=1000, chunk_overlap=30)
docs = text_splitter.split_documents(document)
docs
```

```python
#create embeding
from langchain_community.embeddings import OllamaEmbeddings
from langchain_community.vectorstores import FAISS

embedding = OllamaEmbeddings(model="llama2:7b")
db= FAISS.from_documents(docs, embedding)
```

```python
db
```

```python
#quering 
query = " We have borne with their present government through all these bitter months because of?"
docs= db.similarity_search(query)
docs[0].page_content
```

```python
#retriever class-- act as an interface and can be connected to vactorstore db -- chnage vector db to retriver to get result
retriever= db.as_retriever()
retriever.invoke(query)
docs[0].page_content
```

```python
#similarity search with score
docs_and_score = db.similarity_search_with_score(query)
docs_and_score
```

```python
#use directly as vector
embedding_vector = embedding.embed_query(query)
embedding_vector
```

```python
docs_score = db.similarity_search_by_vector(embedding_vector)
docs_score
```

```python
#saving and loading
db.save_local("faiss_index")
```

```python
#load
new_db = FAISS.load_local("faiss_index", embedding, allow_dangerous_deserialization=True)
```

```python
docs= new_db.similarity_search(query)
```

```python
##chroma DB
from langchain_chroma import Chroma
from langchain_community.document_loaders import TextLoader
from langchain_community.embeddings import OllamaEmbeddings
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
loader = TextLoader("speech.txt")
data = loader.load()

data
```

```python
text_split = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=0)
splits = text_split.split_documents(data)
```

```python
embedding = OllamaEmbeddings(model= "llama2:7b")
vectordb= Chroma.from_documents(documents=splits, embedding=embedding)
```

```python
query = "Just because we fight without rancor and without selfish object"
docs = vectordb.similarity_search(query)
docs[0].page_content
#save
vectordb= Chroma.from_documents(documents=splits, embedding=embedding, persist_directory="./chroma_db")
```
